```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

First, need to check whether BIC/time is consistent at different angles for 116Cd (d,p) from February. This should check that the high cross section at 25 degrees is as a result of the BIC

```python
angles = np.array([10,18,25,31,40])
S1 = np.array([78052,159681,36253,335027,223929])
S3 = np.array([5685,5200,1849,2526,742])
t = np.array([1800,900,300,1800,1800])
bt = (S1-S3)/t
```

```python
aperture_correction = 14.03/7.249
current_correction = 4

#bt[0] = bt[0] * current_correction * aperture_correction
```

```python
plt.plot(angles,bt, 'ro')
plt.xlabel("angle(degrees)")
plt.ylabel("BIC/t")
plt.title("BIC/t vs angle for 116Cd(d,p) bite 1 February")
```

Now we need to do Y/t for l = 0 state, in this case the GS.

```python
Yields = np.array([4753,1857,2076,16495,9047])
sYields = np.array([71,43,46,131,97])
Yt = Yields/t
sYt = sYields/t

Yt[0] = Yt[0] * current_correction * aperture_correction
```

```python
plt.errorbar(angles,Yt, sYt, fmt = 'x', lw = 2)
plt.ylabel("Yield/time (s^-1)")
plt.xlabel("Angle(degrees)")
```

```python
t_a = np.arange(61)
t0xs = np.loadtxt(zero_f)[:,1]
t2xs = np.loadtxt(two_f)[:,1]
```

```python
from scipy.optimize import curve_fit

def objective(angles,txs,N):
    t_a = np.arange(61)
    return txs[np.intersect1d(t_a,angles)] * N

def norm(t_a, txs, a, xs, e):
    p1 = curve_fit(lambda angles, N: objective(angles,txs,N), a, xs, sigma = e)
    return p1[0]
```

```python
Nfeb0 = norm(t_a, t0xs, angles, febxs, sfebxs)
Nmay0 = norm(t_a, t0xs, mangles, mayxs, smayxs)
plt.errorbar(angles,febxs,sfebxs,fmt = 'x', color = 'b')
plt.plot(t_a,Nfeb0 * t0xs, 'b', label = 'Feb')
plt.errorbar(mangles,mayxs,smayxs,fmt = 'x', color = 'r')
plt.plot(t_a,Nmay0 * t0xs, color = 'r', label = 'May')
plt.legend()
plt.text(55,8, '0 keV')
plt.text(22,10,"116Cd(d,p)117Cd")
plt.xlabel('Angles', size = 'large')
plt.ylabel('Differential Cross Section (mb/sr)', size = 'large')

plt.savefig('116_compl0.png')
```

```python
import numpy as np

xs40 = np.array([1.91592018268878, 1.84201887133254, 1.82035048934717, 1.83346218782765, 1.95647017849238])
e40 = np.array([0.030926643970232,0.084975745957218,0.119201201014009,0.067709786160767,0.16658544255482])

xs31 = np.array([3.00053935218303,2.97239972837506,3.01602973725407,3.00800510866697,2.91435661662106])
e31 = np.array([0.035077901309667,0.473761546268121,0.485830922447924,0.094139236991213,0.162770005829802])

xs25 = np.array([4.65283825691474,4.07932572192992,4.25652870562466,4.46276880967201,4.60859708951141])
e25 = np.array([0.123512501348638,0.373750552311289,0.268625319205835,0.243948633166865,0.719344654885103])

xs18 = np.array([3.0793775299859,2.64448716317195,2.72087666126213,2.77992177024867,3.38067489542824])
e18 = np.array([0.118118529912863,0.220776984787618,0.125487346102022,0.107214190435241,0.408177918589954])

```

```python
avg40 = np.average(xs40, weights = 1/e40**2)
err40 = np.sqrt(1/np.sum(1/e40**2))

avg31 = np.average(xs31, weights = 1/e31**2)
err31 = np.sqrt(1/np.sum(1/e31**2))

avg25 = np.average(xs25, weights = 1/e25**2)
err25 = np.sqrt(1/np.sum(1/e25**2))

avg18 = np.average(xs18, weights = 1/e18**2)
err18 = np.sqrt(1/np.sum(1/e18**2))
```

```python
angles_list = [40,31,25,18]
ratios_list = [avg40, avg31, avg25, avg18]
error_list = [err40, err31, err25, err18]
```

```python
import glob
import pandas as pd
files_to_norm = glob.glob('/home/ben/Documents/analysis_code/spectrum_analysis/output/*')

for file in files_to_norm:
    for i,angle in enumerate(angles_list):
        if float(file[75:77]) == float(angle):
            dfg = pd.read_table(file, sep = ' ')
            newfilename = file + '_old'
            dfg.to_csv(newfilename, sep = ' ')
            dfg['CROSS_SECTION'] = dfg['CROSS_SECTION'] * ratios_list[i]
            dfg['ERROR'] = dfg['CROSS_SECTION'] * ((dfg['ERROR']/(dfg['CROSS_SECTION']/ratios_list[i]))**2 + (error_list[i]/ratios_list[i])**2)**0.5
            dfg.to_csv(file, sep = ' ')
            
dfg
```
